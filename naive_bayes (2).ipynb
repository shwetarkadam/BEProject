```
import pandas as pd
import numpy as np
import pickle
```

Import the dataset.

```
df = pd.read_csv("dataset2.csv")

review_text = df.iloc[:, 0]
good = df.iloc[:, 4]
```

```
print(review_text.head())
print(good.head())
```

```
print(len(review_text))
```

Split the data into test and training set.

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(review_text, good, test_size=0.4, random_state=1)
```

Train vectorizer for feature extraction of reviews.

```
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(x_train)
vectorizer.fit(x_test)
```

```
print(vectorizer.vocabulary_)
print(len(vectorizer.vocabulary_))
```

Save the vectorizer

```
with open("vectorizer.pkl", "wb") as f:
  pickle.dump(vectorizer, f)
```

Load the vectorizer

```
with open("vectorizer.pkl","rb") as f:
  vectorizer = pickle.load(f)
```

Do feature extraction of training and test reviews.



```
x_train = vectorizer.transform(x_train)
x_test = vectorizer.transform(x_test)
```

Train the Gaussian Naive Bayes model.

```
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
gnb.fit(x_train.toarray(), y_train)
```

```
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB() 
bnb.fit(x_train.toarray(), y_train) 
  

y_pred = bnb.predict(x_test.toarray()) 
  
from sklearn import metrics 
print("Bernoulli Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred))
```

```
from sklearn.naive_bayes import MultinomialNB 
mnb = MultinomialNB() 
mnb.fit(x_train.toarray(), y_train) 
  

y_pred = mnb.predict(x_test.toarray()) 
  
from sklearn import metrics 
print("Multinomail Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred))
```

```
from sklearn.naive_bayes import ComplementNB 
cnb = ComplementNB() 
cnb.fit(x_train.toarray(), y_train) 
  

y_pred = cnb.predict(x_test.toarray()) 
  
from sklearn import metrics 
print("Complement Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred))
```
